# Distribuição amostral e intervalo de confiança

```python
# Variável com os dados da coluna salário
salarios = dados['SALARIO']
salarios
```

```python
# Variável com a média dos salários
media_amostral = np.mean(salarios)
media_amostral
```

```python
# Variável com o desvio padrão dos salários
desvio_amostral = np.std(salarios)
desvio_amostral
```

```python
# Variável com o nível de confiança 95%
nivel_confianca = 0.95
```

```python
# Variável com a quantidade de salários
tamanho_amostra = len(salarios)
tamanho_amostra
```

```python
# Importando apenas um módulo da biblioteca scipy
from scipy import stats
```

```python
# Variável com o erro médio padrão
erro_padrao = stats.sem(salarios)
erro_padrao
```

```python
# Variável com o intervalo de confiança
intervalo_confianca = stats.t.interval(nivel_confianca, tamanho_amostra-1, loc=media_amostral, scale=erro_padrao)
intervalo_confianca
```

# Featuring engineering

```python
# Retorna "Pessoa Gestora" se o valor de 'gestor' for 1, caso contrário, retorna o valor original de 'nivel'
def preencher_nivel(gestor,nivel):
  if gestor==1:
    return "Pessoa Gestora"
  else:
    return nivel
```

```python
# Cria a coluna 'NOVO NIVEL', preenchendo com "Pessoa Gestora"
dados['NOVO NIVEL'] = dados.apply(lambda x: preencher_nivel(x['GESTOR?'],x['NIVEL']),axis=1)
```

```python
# Classifica a idade em faixas geracionais: X, Millenial, Z ou outra
def determinar_geracao(idade):
  if 39 < idade <= 58:
    return 'Geração X'
  elif 29 < idade <= 39:
    return 'Millenial'
  elif 13 < idade <= 29:
    return 'Geração Z'
  else:
    return 'Outra Geração'
```

# Correlação

```python
# Importando biblioteca
from scipy.stats import chi2_contingency
```

```python
# Calcula e armazena a correlação entre as variáveis contínuas 'IDADE' e 'SALARIO' no DataFrame 'dados'.
correlacao_continua = dados['IDADE'].corr(dados['SALARIO'])
correlacao_continua
```

```python
# Calcula o coeficiente de Cramér para medir a associação entre duas variáveis categóricas
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```

```python
# Calcula o coeficiente de Cramér para verificar a força da associação entre as variáveis 'COR/RACA/ETNIA' e 'NIVEL DE ENSINO' no DataFrame 'dados'
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```

```python
# Cria e mostra uma tabela cruzada (frequência absoluta) entre as variáveis categóricas 'COR/RACA/ETNIA' e 'NIVEL DE ENSINO' no DataFrame 'dados'
tabela_cruzada = pd.crosstab(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
tabela_cruzada
```
